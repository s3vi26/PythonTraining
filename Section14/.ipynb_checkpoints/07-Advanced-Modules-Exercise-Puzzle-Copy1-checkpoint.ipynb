```python
for folder, sub_folders, files in os.walk('C:\\Users\\Sevi\\code\\Complete-Python-3-Bootcamp\\12-Advanced Python Modules\\08-Advanced-Python-Module-Exercise\\Instructions_unzip\\extracted_content'):
    print(f"Currently looking at {folder}")
    print("\n")
    print('The subfolders are: ')
    for sub_fold in sub_folders:
        print(f"Subfolder: {sub_fold}")
    print('\n')
    print('the files are: ')
    for f in files:
        print(f"\t File: {f}")
        the_file = open(f'{folder}\\{f}', 'r+')
        contents = the_file.read()
        the_file.close()
        phone_pattern = re.compile(r'(\d{3})-(\d{3})-(\d{4})')
        results = re.search(phone_pattern, contents)
        if results:
            print("HERE IS THE MATCH!!")
            print(results)
    print("\n")
```
